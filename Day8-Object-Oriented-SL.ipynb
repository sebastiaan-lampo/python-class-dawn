## Object Attributes are Mutable

This means similar to lists if I pass an object to a function by reference (the default) then any changes done to the attributes will affect the original object. If I don't want that, we need to supply a copy of the object to the function instead.

```python

def updateModel(myCar):
    myCar.modelName="Honda"
    
    
myCar=Car()
print(myCar.modelName)
updateModel(myCar)
print(myCar.modelName)
```

To pass an object by the value (instead of by reference) you may use the `copy` library. 

```python
import copy

myCar=Car()
print(myCar.modelName)
updateModel(copy.copy(myCar))
print(myCar.modelName)
```

**SL EXTRA**: When you write a function, clearly document in your docstring whether it changes the arguments in-place (the object should be passed by reference) or not (the object should be a copy and will be returned).  
Remember list.reverse()? That is an in-place function. It does not return a value but does it's work on the object itself.  
By contrast, pandas, the data analysis library uses almost exclusively functions that return a copy.  
If you are unsure, I suggest that you make a copy inside your function. That way there is less risk of making unintentional changes to a mutable argument.

```python
import copy
def add_one_to_first_element(source_list):
    local = copy.copy(source_list)  # Do this for safety unless you have a clear need for in-place change. This way the statements below will function as expected.
    local[0] = local[0] + 1
    return local

a = [0, 1]
b = add_one_to_first_element(a)
print(f"a={a}, b={b}")  # Without copy the result would be a=[1, 1], b=[1, 1]
```

## Class Methods

In the above example, we mainly worked with object attributes. What if I want to print the object without having to call each attribute by itself

```python
myCar=Car()
print(myCar.modelName)
print(myCar.doorsNum)
print(myCar.color)
```

Instead of calling each attribute and print it, I want to call one function that can print for me all these values

```python
class Car:
    def __init__(self,modelName="UnDefined",doorsNum=-1,color="UnDefined"):
        self.modelName=modelName
        self.doorsNum=doorsNum
        self.color=color
    def printAttributes(self):
        print("Model name is ",self.modelName)
        print("Number of doors is ",self.doorsNum)
        print("Car color is ",self.color)
```

```python
myCar=Car()
myCar.printAttributes()
```

A function inside a class is called a method. Currently we have created two methods (`__init__(), printAttributes())

```python
myCar=Car()
print(myCar)
```

Now if I want to print the object as it will only print for the memory address this object is at.

To handle this, we need to override the builtin python `__str__()` method. We replace the `printAttributes()` method with the `__str__()`

```python
class Car:
    def __init__(self,modelName="UnDefined",doorsNum=-1,color="UnDefined"):
        self.modelName=modelName
        self.doorsNum=doorsNum
        self.color=color
    def __str__(self):
        return "\n".join([f"Model name is {self.modelName}", 
                          f"Number of doors is {self.doorsNum}",
                          f"Car color is {self.color}"])
```

```python
myCar=Car()
print(myCar)
```

 # Public vs Private Attributes and Methods

In some programming languages, you can control how widely accessible attributes and methods are. While Python strictly speaking does not allow this, there are still 3 levels you can choose from. These levels apply equally to functions and variables defined in a module or in a class. They are:
- public: Accessible from any other code. Most of your functions should be like this as well as many of your attributes. 
- protected: Should not be used by outside code, but if they need to really access advanced functionality, they can. This is indicated by a single leading `_`
- private: Really for internal use only. Outside code has to jump some hoops to access this. Indicated by a double leading `__`

Your teacher confuses the matters by referring to the dunder methods (e.g. `__init__`) as private. While they are private, their role is really to work with python internals and keywords like you've seen with:
- `car = Car()` <- calls the `__init__`
- `print(car)` <- calls the `__str__`
As you can see from the `dir()` inspection, there are many more.

It's usually safe to make attributes and methods accessible, but there are cases where you would rather not. Examples are if you need to do some validation before changing a variable. Use private attributes and methods responsibly (don't make everything private by default).

```python
class Car:
    def __init__(self,modelName="Tesla",doorsNum=None,color="Red"):
        self.modelName=modelName
        self._doorsNum=doorsNum
        self.__color=color  # SL: For example only, not a good use case for private.
        
    # SL: the following would be a simplistic example of how to make the number of door accessible to an outside user with some added functionality.
    def set_doors_num(self, doors_num):
        if doors_num > 0:  # Don't assign non-sensical numbers
            self._doorsNum = doors_num
            
    def get_doors_num(self):
        return self._doorsNum
        
myCar=Car()        
print(myCar.modelName)

print("\nPlaying with doors:")
print(myCar._doorsNum)  # SL: This works. Protected is still accessible but the _ signals that this is maybe not the right approach
myCar._doorsNum = -3.1415  # SL This also works. It bypasses the check by setting the attribute directly.
print(myCar.get_doors_num()) 
print(myCar.set_doors_num(4))  # SL: The preferred methods that you'd find easily by doing dir(Car)
print(myCar.get_doors_num())  # SL: The preferred methods that you'd find easily by doing dir(Car)

print("\nShowing of private attributes")
try:
    print(myCar.__color)  # This fails. Python hides the private attributes a little better
except AttributeError as e:
    print(f"ERROR: {e}")
print(myCar._Car__color)  # SL: This works. It's a clear sign that you're not supposed to do this unless you really have a need for it. The process is called name mangling which you may now promptly forget.
```

# If everything is an object in Python, why don't we use classes more often?

The answer to this can be unfortunately phrased as a question: why would we? There are many situations in which classes will improve the functionality of your program and provide convenient solutions to problems. However, there are also situations in which using objects will only unnecessarily complicate things. 

At the outset of writing any code to accomplish a task, and as you write the code, it is helpful to ask yourself the following questions:

* How can I store data and variables in my program? Do lists and dictionaries give me the flexibility that I need?
* Is there another existing object that I can use, rather than 'reinventing the wheel'?
* Is there something in the Python standard library or a well-documented external package that does the same thing that I am trying to do?

# Another Example


Lets create a Time class that has three attributes, namely hours, minutes, and seconds.

SL: Note that there is a very well documented python standard library called datetime that does all the functionality below of course ;-)

```python
class Time:
    
    def __init__(self, hours, minutes, seconds):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds
        
T = Time(hours = 1, minutes = 54, seconds = 35)
```

How can we have this time indicate if am or pm? We can add a boolean variable indcating if AM or not

```python
class Time:
    
    def __init__(self, hours, minutes, seconds,isAm):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds
        self.isAm=isAm
        
T = Time(hours = 1, minutes = 54, seconds = 35,True)
```

Now lets create a function that add time to a `Time` class object.  

```python
class Time:
    
    def __init__(self, hours, minutes, seconds,isAm):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds
        self.isAm=isAm
    def add(self, hours, minutes, seconds):
        self.hours += hours
        self.minutes += minutes
        self.seconds += seconds
    def __str__(self):
        return ""+str(self.hours)+":"+str(self.minutes)+":"+str(self.seconds)+" "+("am" if self.isAm else "pm" )
    
T = Time(hours = 1, minutes = 54, seconds = 35,isAm=True)

print(T)
T.add(3,5,15)
print(T)
```
